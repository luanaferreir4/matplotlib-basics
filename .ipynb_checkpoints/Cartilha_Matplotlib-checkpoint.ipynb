## Importação

```python
import matplotlib.pyplot as plt
```

## plot()
Documentação: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib-pyplot-plot

```python
x = [1, 2, 3, 4]  # horizontal
y = [2, 3, 4, 3]  # vertical

plt.plot(x, y)
plt.show()
```


[Link da imagem](https://matplotlib.org/stable/gallery/showcase/anatomy.html)
> Todos os pontos que possuem círculos são editáveis num gráfico.


```python
x = [1, 2, 3, 4]  # horizontal
y = [2, 3, 4, 3]  # vertical

plt.plot(x, y, label='dados')
plt.ylabel('Eixo y')
plt.xlabel('Eixo x')
plt.title('Título do Gráfico')
plt.xticks([0, 2, 4, 6, 8, 9])
plt.yticks([-1, 3, 5, 9, 11])
plt.show()
```

## LINES propriedades 
(_Propriedades das linhas_)
https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_marker

> - Cores : https://matplotlib.org/stable/tutorials/colors/colors.html#specifying-colors
> - Color picker : https://colorpicker.me/#c324b4

- lw = largura

```python
plt.plot(x, y, linestyle='dashed', color='#c324b4', lw='10.0') 
plt.show()
```

# .Scatter()
(Gráfico de dispersão)
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter

```python
plt.scatter(x, y, c='blue')
plt.show()
```

## .bar()
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib-pyplot-bar

```python
plt.bar(x, y)
plt.show()
```

### Nomeando os eixos de um gráfico

## Mudando os tipos de marcadores do gráfico:
https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

```python
plt.plot(x, y, 'sk')  # o ultimo é o marcador 'marker = 's' color = 'k'
plt.show()
```

## Alterando os limites dos gráficos:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html#matplotlib.pyplot.axis

```python
plt.plot(x, y)
# plt.axis(xmin=-1, xmax=20, ymin=-5, ymax=15)
# plt.axis('auto')
# plt.axis('square')
plt.show()
```

### FIGURE
https://matplotlib.org/stable/api/figure_api.html?highlight=sharex#module-matplotlib.figure

### SUBPLOTS
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

> _figsize_ = tamanho da figura em polegadas
>

```python
valoresx = [1, 2, 3, 4]
valoresy = [1, 4, 2, 3]

figura = plt.figure(figsize = (20, 4))
figura.suptitle('Titulo Geral')
figura.add_subplot(131) # 1 grafico, 3 colunas e a posição do gráfico é 1
plt.plot(valoresx, valoresy, label="um dado qualquer")
plt.legend()
plt.title('Gráfico 1')
plt.show(figura)

figura2 = plt.figure(figsize = (20, 4))
figura2.suptitle('Titulo Geral')
figura2.add_subplot(132) # 1 grafico, 3 colunas e a posição do gráfico é 1
plt.plot(valoresx, valoresy, label="um dado qualquer")
plt.legend()
plt.title('Gráfico 2')
plt.show(figura2)
```
